```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('autos.csv')
df
```

```python
cate = df.select_dtypes(include = 'object').columns
cate
```

```python
from sklearn.preprocessing import LabelEncoder
a = LabelEncoder()
for i in cate:
  df[i] = a.fit_transform(df[i])
```

```python
plt.figure(figsize = (10,10))
pl_nu = 1
for i in df.columns:
  if pl_nu <=4:
    plt.subplot(2,2,pl_nu)
    sns.lineplot(df[i])
  pl_nu +=1
plt.show()
```

```python
plt.figure(figsize = (10,10))
pl_nu = 1
for i in df.columns:
  if pl_nu <=4:
    plt.subplot(2,2,pl_nu)
    sns.boxplot(df[i])
  pl_nu +=1
plt.show()
```

```python
plt.figure(figsize =(15,15))

crr = df.corr()
sns.heatmap(crr,annot = True,linewidths=.7)
```
